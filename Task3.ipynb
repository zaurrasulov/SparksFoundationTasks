# Data visualization

```python
# plotting heatmap of correlation

plt.figure(figsize=(16,9))
ax = sns.heatmap(df.corr(),annot = True,cmap = 'viridis')
plt.show()
```

```python
# plotting the scatter plots

sns.pairplot(df)
```

```python
# plotting the bar charts

sns.barplot(df['Region'], df['Profit'])
```

```python
# plotting the bar charts

sns.barplot(df['Ship Mode'], df['Profit'])
```

```python
# plotting the bar charts

sns.barplot(df['Category'], df['Discount'])
```

```python
# plotting the bar charts

sns.barplot(df['Segment'], df['Sales'])
```

```python
# plotting the bar charts

plot = df.groupby(['State'])[['Profit','Discount','Sales']].sum()
plot = plot.sort_values('Discount')
plot[['Discount']].plot(kind="bar",figsize=(10,8),color = 'tomato')
plt.ylabel("Discount")
plt.show()
```

```python
# plotting the distribution of Sales

sns.distplot(df['Sales'])
```

```python
# plotting the distribution of Discount

sns.distplot(df['Discount'])
```

```python
# plotting the distribution of Profit

sns.distplot(df['Profit'])
```
